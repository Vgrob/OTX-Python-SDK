```python
import pandas as pd
from OTXv2 import OTXv2, IndicatorTypes
from pandas.io.json import json_normalize
from datetime import datetime, timedelta
```

```python
otx = OTXv2("458044df4b3238308723bc3df5b8bdea1504e5d0e19ac9e2c4f505fd6ffec067")
```

```python
pulses = otx.getall()
```

```python
data = json_normalize(pulses)
```

```python
data
```

```python
industries = data.industries
```

```python
industries_dict = {}
```

```python
for list_industry in industries:
    for industry in list_industry:
        if industry in industries_dict:
            industries_dict[industry] +=1
        else:
            industries_dict[industry] =1
```
